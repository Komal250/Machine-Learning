```python
# Simple Line plot 
```

```python
# Create data for plotting       # 100
x = np.linspace(0.10,100)
y = np.sin(x)

# Create a simple Line plot 
plt.plot(x,y)

# Add tittle and labels 
plt.title("Simple Line plot of sin(x)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show the plot 
plt.show()
```

```python
# Customize line style , color, and market 
```

```python
x = np.linspace(0,10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot multiple lines with customize
plt.plot(x, y1, color='red', linestyle='--', marker='o', label='sin(x)')
plt.plot(x, y2, color='blue', linestyle='-', marker='x', label='cos(x)')

plt.title("Customize Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

# Add grid and lagend 
plt.grid(True)
plt.legend()

plt.show()
```

```python
# create random data 
x = np.random.rand(50)     # 0 to 1 madhe 50 random decimal number 
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000* np.random.rand(50)

# Create scatter plot 
plt.scatter(x,y, c=colors, s=sizes, alpha=0.9, cmap='viridis')         #alpha = Colors frequency , s= size: x and data point chi size

plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.colorbar()  # add color scal
plt.show() 
```

```python
# Random data for histogram
data = np.random.randn(1000)   # 100 random numbers

# plot histogram 
plt.hist(data, bins=30, color='green', edgecolor='black', alpha=0.7)      # bins = groups cha data 
                                                                          # edgecolor = border cha side cha rang
plt.title("Hestogram Example")                                            # alpha transparency 
plt.xlabel("Data Range")
plt.ylabel("Frequency")

plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots 
plt.figure(figsize=(10,5))

# First subplot 
plt.subplot(1,2,1) # Rows , column, index
plt.plot(x,y1, 'r--')
plt.title("sin(x)")

# Second subplot
plt.subplot(1,2,2)
plt.plot(x,y2,'b-')
plt.title("cos(x)")

plt.suptitle("Multiple Plots Example")  # Overall title
plt.tight_layout()
plt.show()
```

```python
# Q9: Create a 2x2 grid of subplots showing different plots: line, bar, scatter, and histogram.
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,2,figsize=(10, 9))

# Line Chart
x = np.arange(0,10)
y = 2*x + 3
ax[0,0].plot(x, y)
ax[0,0].set_title("Line Chart")
plt.savefig("Line-chart.png")


# Bar Chart
categories = ['A','B','C','D']
values = [10,15,7,12]
ax[0,1].bar(categories, values)
ax[0,1].set_title("Bar Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.savefig("Bar-chart.png")

# Scatter plot 
x = np.random.rand(50)
y = np.random.rand(50)
ax[1,1].scatter(x, y,color='orange')
ax[1,1].set_title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.savefig("ctter-plot.png")

# Histogram
data = np.random.randn(1000)
ax[1,0].hist(data,edgecolor='black')
ax[1,0].set_title("Histogram")
plt.savefig("Histogram.png")

plt.show()
```
