```python
#8. Create a DataFrame and display only values greater than a given number, then visualize them.
data ={ 'Values': [10, 25, 30, 15, 50, 20]}
df= pd.DataFrame(data)

filtered_df = df[df['Values'] > 20]

plt.bar(filtered_df.index, filtered_df['Values'])

plt.title("Values Greater Than 20")
plt.xlabel("Index") 
plt.ylabel("Values")
print("Filtered Dataframe: \n", filtered_df)
plt.show()
```
