```python
# Importing the numpy Library
import numpy as np
print("Numpy Version:", np.__version__)
```

```python
# Create a 1D array
arr1 = np.array([10,20,30,40,50])
print("1D Array:\n", arr1)

# Create a 2D aaray
arr2 = np.array([[1,2,3], [4,5,6]])
print("\n2D Array:\n", arr2)
print("\nDimensions of arr3:", arr2.ndim)

# Create a 3D array
arr3 = np.array([[[1,2], [3,4]],[[5,6],[7,8]]])
print("\n 3D Array:\n", arr3)

# Check array dimensions 
print("\n Dimensions of arr3:", arr3.ndim)
```

```python
# Array Properties 
```

```python
print("\nArray Shape (rows, columns):", arr2.shape)
print("Array Size (total elements):", arr2.size)
print("Array Data Type:", arr2.dtype)
```

```python
zeros = np.zeros((2,3))  # 2x3 matrix filled with zeros 
ones = np.ones((3,2))    # 3x2 matrix filled with ones
eye = np.eye(3)          # 3x3 indentity matrix 
rand = np.random.rand(2,3) # random values between 0 and 1
arange_arr = np.arange(0,10,2) # Even numbers from 0 to 8 
linspace_arr = np.linspace(0,1,5)  # 5 equally spced values between 0 and 1

print("\nZeros Array\n", zeros)
print("\nOnes Array:\n", ones)
print("\nIndentity Matrix:\n", eye)
print("\nRandom Array:\n", rand)
print("\nArange Array:\n", arange_arr)
print("\nLinspace Array:\n", linspace_arr)
```

```python
# Array Indexing and Slicing 
```

```python
arr= np.array([10,20,30,40,50,60])
print("\nOriginal Array:", arr)
print("First Element:", arr[0])
print("Last Element:", arr[-1])
print("Slice [1:4]:", arr[1:4])
print("Every 2nd Element:", arr[::2])
print("Every 2nd Element:", arr[1::2])

# 2D Array slicing 
arr2d = np.array([[10,20,30], [40,50,60],[70,80,90]])
print("\n2D Array:\n", arr2d)
print("Element at [1,2]:", arr2d[1,2])
print("First two rows and first two columns:\n", arr2d[:2, :2])
```

```python
# Array Operations (Element - wise)
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print("\nAddition:", a+b)
print("Subtraction:", a-b)
print("Multiplication:", a*b)
print("Division:", a/ b)
print("Power:", a**2)
```

```python
# Statistical and Mathmatical Functions
```

```python
arr = np.array([10,20,30,40,50])

print("\nSum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Minimun:", np.min(arr))
print("Maximum:", np.max(arr))
```

```python
# Reshaping and Flattening Arrays
```

```python
arr = np.arange(1,13)
print("\nOriginal 1D Array:", arr)

reshaped = arr.reshape(3,4)
print("Reshaped to 3x4 Matrix:\n", reshaped)

# reshpaed1 = arr.reshape(5,5)
#print("Reshaped to 3x4 Matrix:\n", reshaped1)

flattened = reshaped.flatten()
print("Flattened Back to 1D:", flattened)
```

```python
# Concatenation and Stacking 
```

```python
x = np.array([[1,2], [3,4]])
y = np.array([[5,6],[7,8]])

print("\nHorizontal Stack:\n", np.hstack((x,y)))
print("Vertical Stack:\n", np.vstack((x,y)))
```

```python
# Boolean Indexing and Filtering 
```

```python
arr = np.array([5,10,15,20,25,30])
filter_arr = arr > 15
print("\nFilter Boolean Array:", filter_arr)
print("Filtered Values (arr > 15):", arr[filter_arr])
```

```python
print("\nRandom Integer between 1-100:", np.random.randint(1, 100))
print("Random Float:", np.random.random())
print("Random 3x3 Matrix:\n", np.random.randint(10,50, size=(3,3)))

# print random 5x6 float matrix
```

```python
# Linear Alegebra Operations 
```

```python
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])

print("\n<Matrix 1:\n", mat1)
print("Matrix 2:\n", mat2)
print("\n<Matrix Multiplication:\n", np.dot(mat1,mat2))
print("Transpose of Matrix 1:\n", mat1.T)
print("Determinant of Matrix 1:", np.linalg.det(mat1))
print("Inverse of Matrix 1:\n", np.linalg.inv(mat1))
```

```python
# Copy vs view (memory concep)
```

```python
original = np.array([1,2,3,4])
view_arr = original.view()
copy_arr = original.copy()

original[0] = 99

print("\nOriginal Array:", original)
print("View Array (Changes refelected):", view_arr)
print(" Copy Array (Independent):", copy_arr)
```
