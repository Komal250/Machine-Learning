```python
# 6. Create a random 3x3 matrix, find its transpose, determinant, and inverse.
import numpy as np

matrix = np.random.randint(1, 10, (3, 3))

transpose = matrix.T

determinant = np.linalg.det(matrix)

if determinant != 0:
    inverse = np.linalg.inv(matrix)
else:
    inverse = "Inverse does not exist"

print("Matrix:\n", matrix)
print("Transpose:\n", transpose)
print("Determinant:", determinant)
print("Inverse:\n", inverse)
```

```python
#7. Generate a 5x5 array of random integers between 10 and 100, and filter out values greater than 50
import numpy as np

arr = np.random.randint(10, 101, (5, 5))

filtered = arr[arr > 50]

print("5x5 Array:\n", arr)
print("Values Greater Than 50:\n", filtered)
```

```python
#8. Create two 1D arrays and horizontally and vertically stack them using NumPy functions
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

horizontal = np.hstack((a, b))

vertical = np.vstack((a, b))

print("Horizontal Stack:", horizontal)
print("Vertical Stack:\n", vertical)
```

```python
#9. Create an array of 10 random numbers and replace all values greater than the mean with the mean value
import numpy as np

arr = np.random.randint(1, 100, 10)

mean = np.mean(arr)

modified = np.where(arr > mean, mean, arr)

print("Original Array:", arr)
print("Mean:", mean)
print("Modified Array:", modified)
```

```python
#10. Create an array of 20 consecutive numbers and split it equally into 4 sub-arrays
import numpy as np

arr = np.arange(1, 21)

split_arrays = np.split(arr, 4)

print("Original Array:", arr)

for i, sub_array in enumerate(split_arrays, 1):
    print(f"Sub-array {i}:", sub_array)
```
