```python
# 3. Write a Python program to tokenize a given paragraph into sentences using sent_tokenize()
from nltk.tokenize import sent_tokenize
text = "Komal is learning Natural Language Processing. I lives in Shrirampur"
sentences = sent_tokenize(text)
print("Sentence Tokenization:\n", sentences)
```

```python
# 4. Tokenize the same paragraph into individual words using word_tokenize().
from nltk.tokenize import word_tokenize
words = word_tokenize(text)
print("Eord Tokenization:\n", words)
```

```python
# 5. Remove stopwords from a given text using NLTK stopwords list.
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words]

print("Filtered words:", filtered_words)
```

```python
# 6. Perform stemming on a list of words using the PorterStemmer and print the output.
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in filtered_words]

print("Stemmed words:\n", stemmed_words)
```

```python
# 7. Perform lemmatization on the same list of words using WordNetLemmatizer and compare results with stemming
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]

print("Lemmatized words:\n", lemmatized_words)
```

```python
# 8. Apply Part of Speech (POS) tagging to a given sentence and print each word with its tag.
from nltk import pos_tag 
print(pos_tag(words))
```

```python
# 9. Perform frequency distribution on a paragraph and print the top 10 most frequent words.
from nltk import FreqDist

freq_dist = FreqDist(filtered_words)
print("Words Frequency:\n")
for word, freq in freq_dist.most_common(5):  
    print(f"{word}: {freq}")
```

```python
# 10. Plot a graph of the top 15 most frequent words using FreqDist().plot().
freq_dist.plot(15)
```

```python
# 12. Perform Named Entity Recognition (NER) on a text containing names of people, places, and organizations.
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

text = "Komal Shinde lives in Shrirampur and Leaning data analysis."

words = word_tokenize(text)

pos_tags = pos_tag(words)
ner_tree = ne_chunk(pos_tags)
print("Named Entity Recognition Output:\n")
print(ner_tree)
```

```python
# 13. Create a function that accepts text input and returns a list of nouns and verbs separately using POS tagging.
text = "Komal Shinde lives in Shrirampur and Leaning data analysis."

def get_nouns_and_verbs(text):
    words = word_tokenize(text)
    tags = pos_tag(words)

    nouns = []
    verbs = []

    for word, tag in tags:
        if tag.startswith("NN"):
            nouns.append(word)
        elif tag.startswith("VB"):
            verbs.append(word)

    return nouns, verbs
    
nouns, verbs = get_nouns_and_verbs(text)

print("Nouns:", nouns)
print("Verbs:", verbs)
```

```python
# 14. Count the number of adjectives and adverbs present in a given text.
text = "She is a very smart and highly intelligent girl."

words = word_tokenize(text)
tags = pos_tag(words)

adj = sum(1 for w,t in tags if t.startswith('JJ'))
adv = sum(1 for w,t in tags if t.startswith('RB'))

print(adj, adv)
```

```python
# 15. Extract all bigrams (two-word combinations) from a sentence using nltk.bigrams().
# Import required libraries
import nltk
from nltk.util import bigrams
from nltk.tokenize import word_tokenize

nltk.download('punkt')

sentence = "Natural language processing is very interesting and useful"
words = word_tokenize(sentence)
bigram_list = list(bigrams(words))

print("Bigrams:")
for bg in bigram_list:
    print(bg)
```

```python
# 15. Compare and print the difference between stemming and lemmatization for a list of 10 verbs.
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet

# Download required datasets
nltk.download('wordnet')
nltk.download('omw-1.4')

# List of 10 verbs
verbs = [
    "running",
    "playing",
    "studies",
    "went",
    "eating",
    "driving",
    "written",
    "swimming",
    "talked",
    "flying"
]

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()


print("{:<12} {:<12} {:<12}".format("Original", "Stemmed", "Lemmatized"))
print("-" * 36)

for verb in verbs:
    stemmed = stemmer.stem(verb)
    lemmatized = lemmatizer.lemmatize(verb, pos='v')  # pos='v' for verb
    
    print("{:<12} {:<12} {:<12}".format(verb, stemmed, lemmatized))
```
