```python
# Q1: Create a simple line plot of y = 2x + 3 for x values ranging from 0 to 10.
x = np.arange(0,11)
y = 2*x + 3 

plt.plot(x,y)
plt.title(" Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
plt.savefig("line-chart.png")
```

```python
# Q2: Plot sin(x) and cos(x) on the same graph with different colors, line styles, and labels, and display a legend
x = np.linspace(0,10,100)

plt.plot(x, np.sin(x), linestyle='--', label='sin(x)')
plt.plot(x, np.cos(x), linestyle='-', label='cos(x)')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customize Line Plot")
plt.legend()
plt.show()
plt.savefig("Customize-plot.png")
```

```python
# Q7: Create two subplots in one figure – one showing a sine wave and the other showing a cosine wave.
x = np.linspace(0,10,100)

plt.subplot(1,2,1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x, np.sin(x))


plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.xlabel("X-axis")
plt.suptitle("Multiple Plots Example") 
plt.show()
```

```python
# Q9: Create a 2x2 grid of subplots showing different plots: line, bar, scatter, and histogram.
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,2,figsize=(10, 9))

# Line Chart
x = np.arange(0,10)
y = 2*x + 3
ax[0,0].plot(x, y)
ax[0,0].set_title("Line Chart")


# Bar Chart
categories = ['A','B','C','D']
values = [10,15,7,12]
ax[0,1].bar(categories, values)
ax[0,1].set_title("Bar Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


# Scatter plot 
x = np.random.rand(50)
y = np.random.rand(50)
ax[1,1].scatter(x, y,color='orange')
ax[1,1].set_title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


# Histogram
data = np.random.randn(1000)
ax[1,0].hist(data,edgecolor='black')
ax[1,0].set_title("Histogram")

plt.show()
```
